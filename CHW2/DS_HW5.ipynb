```python
import pandas as pd
from persiantools.jdatetime import JalaliDateTime

def extract_date_parts(persian_date):
    if isinstance(persian_date, str):
        persian_date = persian_date.split('+')[0]  # Remove timezone if present

    try:
        jalali_date = JalaliDateTime.strptime(persian_date, '%Y-%m-%d %H:%M:%S.%f')
        return jalali_date.year, jalali_date.month, jalali_date.day
    except Exception as e:
        print(f"Error parsing date: {persian_date}, {e}")
        return None, None, None  # Return None values if conversion fails

# Apply function to extract year, month, and day
df[["Creation Year", "Creation Month", "Creation Day"]] = df["Creation Date"].apply(lambda x: pd.Series(extract_date_parts(x)))
df[["Finish Year", "Finish Month", "Finish Day"]] = df["Finish Date"].apply(lambda x: pd.Series(extract_date_parts(x)))

# Function to calculate days passed
def calculate_days_passed(row):
    try:
        start_date = JalaliDateTime(row["Creation Year"], row["Creation Month"], row["Creation Day"])
        end_date = JalaliDateTime(row["Finish Year"], row["Finish Month"], row["Finish Day"])
        return (end_date - start_date).days
    except Exception as e:
        print(f"Error calculating days passed: {e}")
        return None

# Apply function to compute days passed
df["Days Passed"] = df.apply(calculate_days_passed, axis=1)

# Save or display result

df[["Creation Date", "Finish Date", "Days Passed"]]
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_selection import SelectKBest, mutual_info_regression

# Convert categorical columns to numerical using Label Encoding
label_encoders = {}
for col in df.select_dtypes(include=['object']).columns:  # Select categorical columns
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col].astype(str))  # Convert to numerical
    label_encoders[col] = le  # Store encoder for future use

# Fill missing values (if any)
df = df.fillna(df.median())

# Define features (X) and target (y)
target_column = "Price"  # Change this if the target is different
X = df.drop(columns=[target_column])
y = df[target_column]

# Feature selection using Mutual Information
feature_selector = SelectKBest(score_func=mutual_info_regression, k='all').fit(X, y)

# Create a DataFrame with feature scores
feature_scores = pd.DataFrame({'Feature': X.columns, 'Score': feature_selector.scores_})
feature_scores.sort_values(by=['Score'], ascending=False, inplace=True)

# Plot top 20 features
plt.figure(figsize=(16, 8))
feature_scores.plot(x='Feature', y='Score', kind='barh', legend=False, figsize=(16, 8), color='skyblue')
plt.xlabel("Mutual Information Score")
plt.ylabel("Features")
plt.title("Features based on Mutual Information")
plt.gca().invert_yaxis()  # Invert y-axis for better visualization
plt.grid(axis="x", linestyle="--", alpha=0.7)
plt.show()


# Display top feature scores
feature_scores
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, mean_absolute_percentage_error

# Select relevant columns
selected_columns = [
    'Parent Service Category', 'Service Name', 'Finish Date',
    'Order Location', 'Average Price by Location', 'Price'
]
df_regression = df[selected_columns]

# Separate features and target
X = df_regression.drop('Price', axis=1)
y = df_regression['Price']

# Normalize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.25, random_state=42
)

# Train regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Predict on test set
y_pred = model.predict(X_test)

# Evaluation metrics
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
mape = mean_absolute_percentage_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Print metrics
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Mean Absolute Percentage Error (MAPE): {mape:.2f}")
print(f"R2 Score: {r2:.2f}")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    precision_score, recall_score, f1_score, classification_report
)
import numpy as np

# Define features and target
features = [
    'No. Previous Orders by Customer', 'Customer Join Date', 'Parent Service Category',
    'Service Name', 'Order Status', 'Order City', 'Order Region',
    'Device', 'Price', 'Discount', 'Quoted Task?'
]
target = 'Revised Previous Orders'

# Extract features and target
X = df[features]
y = df[target]

# Normalize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.9, random_state=42
)

# Train classifier
clf = RandomForestClassifier(random_state=42)
clf.fit(X_train, y_train)

# Predict
y_pred = clf.predict(X_test)

# Precision and Recall for each class
class_labels = np.unique(y_test)
precision_per_class = precision_score(y_test, y_pred, average=None, labels=class_labels)
recall_per_class = recall_score(y_test, y_pred, average=None, labels=class_labels)

# Print per-class precision and recall
print("Precision per class:")
for label, score in zip(class_labels, precision_per_class):
    print(f"  Class {label}: {score:.2f}")

print("\nRecall per class:")
for label, score in zip(class_labels, recall_per_class):
    print(f"  Class {label}: {score:.2f}")

# F1-scores with different averaging
f1_macro = f1_score(y_test, y_pred, average='macro')
f1_weighted = f1_score(y_test, y_pred, average='weighted')
f1_micro = f1_score(y_test, y_pred, average='micro')

print(f"\nF1 Score (macro): {f1_macro:.2f}")
print(f"F1 Score (weighted): {f1_weighted:.2f}")
print(f"F1 Score (micro): {f1_micro:.2f}")

# Optional: full classification report
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

# 10 Points: Suppose we have a multi-label classification problem in the field of football, where each sample (player) can belong to some of 4 classes that we have:
* Class 1: The player has played for the national team before
* Class 2: The player had previous history of heart problems
* Class 3: The player had knee injuries before
* Class 4: The player has been the captain of the team in the past

In a multi-label classification problem like this—where each player can belong to any subset of the 4 classes (not just one)—we need a metric that evaluates the correctness across all labels for each instance.

Recommended Accuracy Metric: Subset Accuracy (Exact Match Ratio)
Definition:
Subset Accuracy (also known as Exact Match Ratio) checks whether the entire set of predicted labels exactly matches the true set of labels for each sample. It’s a very strict metric.

Why it's suitable here:
In your case, all four classes are important, and predicting all of them correctly matters, especially in a sensitive context like health or leadership history.

Subset accuracy ensures that the classifier gets every label right per player, which is critical when misclassifying even one aspect (like missing a heart problem) could be serious.

It avoids the problem of a model being rewarded for partially correct predictions.

Example:
Suppose the true labels for a player are:
[1, 0, 1, 1] (has played for national team, no heart problem, had knee injury, was captain)

And your model predicts:
[1, 0, 1, 0] (missed the captain history)

Subset Accuracy would treat this prediction as incorrect (score = 0) because not all labels match.

This is helpful in high-stakes domains where all label components matter.



```python
import numpy as np
from sklearn.metrics import accuracy_score

# Step 1: Generate dummy true labels for 100 players (each has 4 binary labels)
np.random.seed(42)
n_samples = 100
n_labels = 4

# Each row is a player, each column is a class (1: has it, 0: doesn't)
y_true = np.random.randint(0, 2, size=(n_samples, n_labels))

# Step 2: Generate random predictions (like from a bad model or for testing)
y_pred = np.random.randint(0, 2, size=(n_samples, n_labels))

# Step 3: Compute Subset Accuracy
subset_accuracy = accuracy_score(y_true, y_pred)

print(f"Subset Accuracy (Exact Match Ratio): {subset_accuracy:.2f}")
```

a softer metric (for evaluation or model development), consider:
Hamming Loss: Measures the fraction of labels that are incorrectly predicted. Good for tracking per-label performance.

Macro/Micro-Averaged F1 Score: Useful when class imbalance is a concern, especially if some classes are rarer (like heart problems).

```python
from sklearn.metrics import hamming_loss, f1_score

print("Hamming Loss:", hamming_loss(y_true, y_pred))
print("Micro F1 Score:", f1_score(y_true, y_pred, average='micro'))
print("Macro F1 Score:", f1_score(y_true, y_pred, average='macro'))
```
